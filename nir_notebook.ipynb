```python
all_logs = pd.read_csv('./logon+device+file_order_by_datetime.csv')
#all_logs['day'] = pd.to_datetime(all_logs['date'])
all_logs
```

```python
new_logs = all_logs[(all_logs.datetime >= '2009-02-04') & (all_logs.datetime < '2010-01-23')]
```

```python
d = new_logs.to_dict()
```

```python
pd.DataFrame(d)
```

```python
logon_logs = new_logs[new_logs['type'] == 'logon']
```

```python
logon_logs['date'] = pd.to_datetime(logon_logs['datetime']).dt.date
logon_logs['nday'] = pd.to_datetime(logon_logs['date']).dt.day_of_week
#logon_logs['is_weekend'] = 0
logon_logs['is_weekend'] = logon_logs['nday'].apply(lambda x: 1 if x in [5,6] else 0)
```

```python
logon_logs
```

```python
dl2 = pd.DataFrame({'number_of_logs': logon_logs.groupby(['user','is_weekend','date']).size()}).reset_index()
dl2
```

```python
dl4 = pd.DataFrame({'sum_of_logs': logon_logs.groupby(['user','is_weekend']).size()}).reset_index()
dl4 = dl4[dl4['sum_of_logs'] != 0].reset_index(drop=True)
dl4
```

```python
dl3 = pd.DataFrame({'mean': dl2.groupby(['user','is_weekend'])['number_of_logs'].agg('mean')}).reset_index()
dl3
```

```python
feature101 = pd.merge(dl2, dl4, on=["user",'is_weekend'])
#feature101 = pd.merge(feature101, dl3, on=["user",'is_weekend'])
feature101
```

```python
# фича101 с подсчетом среднего отклонения

feature101['freq'] = 0.0
feature101['freq'] = feature101['number_of_logs'].to_numpy() / feature101['sum_of_logs'].to_numpy()

feature101['mean_freq'] = 0.0
feature101['mean_freq'] = feature101['mean'].to_numpy() / feature101['sum_of_logs'].to_numpy()

feature101['mean_dev'] = feature101['freq'] - feature101['mean_freq']
feature101['mean_dev'] = feature101['mean_dev'].apply(lambda x: x if x > 0 else 0)

lst = [0] + list(range(2,9))
feature101 = feature101[feature101.columns[lst]]
feature101
```
